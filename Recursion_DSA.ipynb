# **Factorial for DSA**

#Factorial using recursion

```python
num = int(input("Enter a number: "))
def factorial(num):
  if num == 0 or num ==1:

    return 1

  return num*factorial(num-1)

print(f"The factorial of {num} is {factorial(num)}")
```

# Normal recursion(Head recursion)

```python
def func(count_val):
  if count_val == 4:
    return
  print("Soy")
  count_val += 1
  func(count_val)

count = 0
func(count)
```

## Tail Recursion

```python
count = 0
def func():
  global count # Declare 'count' as global to modify it
  if count == 4:
    return
  count += 1
  func()
  print("soy")

func() # Call the function to execute it
```

# Head and Tail recursion

```python
def head_rec(n):
  if n == 0:
    return
  head_rec(n-1)
  print(n-1)

head_rec(5)

```

```python
def Tail_rec(n):
  if n == 0:
    return
  print(n)
  Tail_rec(n-1)

Tail_rec(5)
```

# **# Recursion using parameter**

## Simple factorial

```python
def factorial(n):
  if n == 0 or n == 1:
    return n
  return n*factorial(n-1)
factorial(1)
```

##Print X in N no of times

```python
def fun(x,n):
  if n == 0:
    return
  print(x)
  fun(x,n-1)
fun(15,4)
```

##Print 1 to n using recursion

```python
def fun(i,n):
  if i>n:
    return
  print(i)
  fun(i+1,n)

fun(1,4)
```

## Print N to 1 using recursion(backtracking)

```python
def fun(i,n):
  if i > n:
    return
  fun(i+1,n)
  print(i)

fun(1,4)
```

## Sum of first n number

```python
def sum(n):
  if n == 0:
    return 0
  return n + sum(n-1)
sum(3)
```

## sum of n number using tail

```python
def tail_sum(n,total):
  if n == 0:
    return total
  return tail_sum(n-1,total+n)
tail_sum(4,0)
```

## Sum of n natural number


```python
def nat(current_sum, i, n):
  if i > n:
    print(current_sum)
    return
  nat(current_sum + i, i + 1, n)

nat(0, 1, 4)
```

# **#Reverse the Array using recursion**

```python
num =[1,2,3,4,5,6,7,8,9]

def func(arr,left,right):
  if left >= right:
    return
  arr[left] , arr[right] = arr[right], arr[left] # Corrected assignment operator
  func(arr,left + 1, right - 1)

def rev_arr(arr,left,right):
  func(arr,left,right) # Corrected function call from self.func to func
  return arr

print("Original array:", num)
reversed_num = rev_arr(num, 0, len(num) - 1)
print("Reversed array:", reversed_num)
```

# **# Check whether string is Palindrom or not**

```python
def palindrom(s,left,right):
  if left >= right:
    return True
  if s[left] != s[right]:
    return False
  return palindrom(s,left+1,right-1)

text = "mat"
print(palindrom(text,0,len(text)-1))

```

#By using Loop

```python
def palindrom(s):
  s = s.lower()

  left = 0
  right = len(s)-1


  while left < right:
    if s[left] != s[right]:
      return False
    left += 1
    right -= 1

  return True

text = 'madam'
print(palindrom(text))
```

# **# Fibonacci Series**

```python
def fibonacci(n):
  if n == 0 or n == 1:
    return n
  return fibonacci(n-1) + fibonacci(n-2)
n = 7
print(fibonacci(n))
```

# **#Using Loop**

```python
def fibonacci(n):
  a,b = 0,1
  for i in range(n):
    a,b = b, a+b
  return a

n = 7
print(fibonacci(n))
```
